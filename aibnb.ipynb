```python
#讀取檔案
import glob
flist = glob.glob("airbnb/*.csv")
flist
```

```python
import pandas as pd
# Concat第一種寫法
df = None
for fn in flist:
    temp = pd.read_csv(fn, encoding = "utf-8")
    if df is None:
        # 第一個csv -> df設定成第一個
        df = temp
    else:
        # 第二個csv以後 -> 嫁接到原本的csv
        df = pd.concat([df,temp], axis = 0)
df
```

```python
# 先準備等一下要接所有csv的list
alldf = []
for fn in flist:
    # 每讀一個, 就把它丟進list
    temp = pd.read_csv(fn, encoding="utf-8")
    alldf.append(temp)
# 最後一次concat
df2 = pd.concat(alldf, axis=0)
df2
```
